```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
amber = cv2.imread('img/amberTurd.jpg')
amber1 = cv2.cvtColor(amber, cv2.COLOR_BGR2RGB)

plt.figure(figsize = (15,15))
plt.imshow(amber1)
```

```python
amber3 = amber1.copy()
font = cv2.FONT_HERSHEY_PLAIN
text = 'Who are you'
textsize = cv2.getTextSize(text,font,1,5)[0]
textX = (amber3.shape[1] - textsize[0])/2
textY = (amber3.shape[0] + textsize[1])/2
textX=round(textX)
textY=round(textY)
cv2.putText(amber3,text,(textX,textY),font,1,(0,0,0),2)
plt.figure(figsize = (15,15))
plt.imshow(amber3)
```

```python
text2 = """Amber Turd"""


cv2.rectangle(amber3, (200,0),(500,25),(0,0,0),cv2.FILLED)
cv2.putText(amber3,'Amber Heard',(200,25),font,2.5,(255,255,255),2)
cv2.rectangle(amber3, (200,350),(500,381),(0,0,0),cv2.FILLED)
cv2.putText(amber3,text2,(200,381),font,3,(255,255,255),2)

plt.figure(figsize = (15,15))
plt.imshow(amber3)
```
