```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
image_ano = cv2.imread('img/2.jpg')
img = cv2.cvtColor(image_ano, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (15,15))
plt.imshow(img)
```

```python
image_ano2 = img.copy()
#flower
cv2.rectangle(image_ano2,(450,90),(1220,750),(255,0,0),4)
cv2.rectangle(image_ano2,(450,40),(650,90),(255,0,0),cv2.FILLED)
cv2.putText(image_ano2,'flower',(450,90),cv2.FONT_HERSHEY_COMPLEX,1.5,(0,0,0),2)
#curtain
cv2.rectangle(image_ano2,(150,0),(1600,200),(200,0,90),4)
cv2.rectangle(image_ano2,(150,200),(320,250),(200,0,90),cv2.FILLED)
cv2.putText(image_ano2,'curtain',(150,250), cv2.FONT_HERSHEY_DUPLEX,1.5,(0,0,0),2)
#railing
cv2.rectangle(image_ano2,(0,400),(1600,800),(255,255,0),4)
cv2.rectangle(image_ano2,(1400,350),(1600,400),(255,255,0),cv2.FILLED)
cv2.putText(image_ano2,'railing',(1400,400), cv2.FONT_HERSHEY_DUPLEX,1.5,(0,0,0),2)
#window
cv2.rectangle(image_ano2,(150,320),(350,400),(0,255,0),4)
cv2.rectangle(image_ano2,(0,320),(150,370),(0,255,0),cv2.FILLED)
cv2.putText(image_ano2,'window',(0,370), cv2.FONT_HERSHEY_DUPLEX,1,(0,0,0),2)

plt.figure(figsize = (15,15))
plt.imshow(image_ano2)
```
